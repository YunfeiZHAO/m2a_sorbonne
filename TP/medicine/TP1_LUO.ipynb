# **Decision tree**

```python
clf_tree = tree.DecisionTreeClassifier()
clf_tree = clf_tree.fit(data_diabetes, classes_diabetes)
```

```python
feature_names = ['age', 'hba1c', 'insuline taken', 'other drugs taken']
classes = ['DR','NDR']
dot_data = tree.export_graphviz(clf_tree, out_file=None,
                                             feature_names=feature_names,
                                             class_names=classes,
                                             filled=True,
                                             rounded=True,
                                             special_characters=True)

graph = graphviz.Source(dot_data)
graph.render("diabetes remission")
```

# **Diabetes Remission score**

```python
def predict_Remission(data):
  score=np.zeros(len(data))
  pred=np.zeros(len(data))
  score[data[0]>=60]+=3
  score[(data[0]>=50)&(data[0]<60)]+=2
  score[(data[0]>=40)&(data[0]<50)]+=1

  score[(data[1]>=6.5)&(data[1]<9)]+=2
  score[(data[1]>=7)&(data[1]<9)]+=4
  score[data[1]>=9]+=6

  score[data[3]==1]+=3
  score[data[2]==1]+=10
  pred[score>=7]=1
  return (pred)
```

```python
predict_Remission(data_diabetes)
```

```python
200-(predict_Remission(data_diabetes)==class_dia).sum()
```

By the state-of-the-art clinical score DiaRem model, we predict the training set with 63 errors for 200 data.

Among the 3 models we present above, it seems that the decision tree perform the best, with zero error in the prediction, but it is more likely because of overfitting, if we have a testing set with data do not trained in the models, we might get a different result.

# **With splitting the set**

Now, let's have a look of the performence of the models when we split the data set into a training set and a testing set, the training set contain 160 patients' data and the testing set contain 40 patients' data.

```python
clf_tree_split = tree.DecisionTreeClassifier()
clf_tree_split = clf_tree.fit(data_diabetes[20:180], classes_diabetes[20:180])
```

```python
40-(sum(clf_tree_split.predict(data_diabetes[0:20])==1)+sum(clf_tree.predict(data_diabetes[180:])==0))
```

```python
clf_RF_split = RandomForestClassifier(max_depth=2, random_state=0)
clf_RF_split.fit(data_diabetes[20:180], classes_diabetes[20:180])
```

```python
40-(sum(clf_RF_split.predict(data_diabetes[0:20])==1)+sum(clf_RF_split.predict(data_diabetes[180:])==0))
```

```python
40-(sum(predict_Remission(data_diabetes[0:20])==1)+sum(predict_Remission(data_diabetes[180:])==1))
```

We can see that we can not have a fully correct prediction with decision tree now, and the prediction error of random forest is near to the prediction decision tree model. Among them, the clinical score DiaRem model predict the best.
